7. create a column named: "decision", If wheel-base is less than 100 and length is less than 175 then 
"don't buy" else "buy"

```python
df['decision'] = df.apply(lambda x: 'don\'t buy' if (x['wheel-base'] < 100) and (x['length'] < 175) else 'buy', axis=1)

# df.to_csv('automobile_data_updated.csv', index=False)
df.head()
```

8. Select only cars whose decision is "buy"; find average price based on company of car

```python
df_av_p = pd.DataFrame(df[df['decision']=='buy'].groupby('company')['price'].mean().round(2).reset_index())
df_av_p
```
